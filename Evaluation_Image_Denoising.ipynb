```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Path to the text file
file_path = '/content/GAN_LossesAfterEpoch_L4_WGP_Transpose.txt'

# Initialize lists to hold the data
steps, d_loss, g_loss, d_loss_avg, g_loss_avg, d_loss_val, g_loss_val, psnr, psnr_avg, psnr_val = [], [], [], [], [], [], [], [], [], []

# Read and parse the file
with open(file_path, 'r') as file:
    for line in file:
        # Extract data from each line
        parts = line.strip().split()
        steps.append(int(parts[2]))
        d_loss.append(float(parts[4]))
        g_loss.append(float(parts[6]))
        d_loss_avg.append(float(parts[8]))
        g_loss_avg.append(float(parts[10]))
        d_loss_val.append(float(parts[12]))
        g_loss_val.append(float(parts[14]))
        psnr.append(float(parts[16]))
        psnr_avg.append(float(parts[18]))
        psnr_val.append(float(parts[20]))

# Create a DataFrame
df = pd.DataFrame({
    'step': steps,
    'd_loss': d_loss,
    'g_loss': g_loss,
    'd_loss_avg': d_loss_avg,
    'g_loss_avg': g_loss_avg,
    'd_loss_val': d_loss_val,
    'g_loss_val': g_loss_val,
    'psnr': psnr,
    'psnr_avg': psnr_avg,
    'psnr_val': psnr_val
})

# Display the DataFrame
print(df)
```

```python
df
```

```python
# Convert steps to epochs
df['step_normalized'] = df['step'] / 388

# Plotting
plt.figure(figsize=(15, 10))

# Plot g_loss_avg vs g_loss_val
plt.subplot(3, 1, 1)
plt.plot(df['step_normalized'], df['g_loss_avg'], label='g_loss_avg', color='blue')
plt.plot(df['step_normalized'], df['g_loss_val'], label='g_loss_val', color='orange')
plt.xlabel('Steps (normalized)')
plt.ylabel('Generator Loss')
plt.legend()
plt.title('Generator Losses (Average and Validation)')

# Plot d_loss_avg vs d_loss_val
plt.subplot(3, 1, 2)
plt.plot(df['step_normalized'], df['d_loss_avg'], label='d_loss_avg', color='blue')
plt.plot(df['step_normalized'], df['d_loss_val'], label='d_loss_val', color='orange')
plt.xlabel('Steps (normalized)')
plt.ylabel('Discriminator Loss')
plt.legend()
plt.title('Discriminator Losses (Average and Validation)')

# Plot psnr_avg vs psnr_val
plt.subplot(3, 1, 3)
plt.plot(df['step_normalized'], df['psnr_avg'], label='psnr_avg', color='blue')
plt.plot(df['step_normalized'], df['psnr_val'], label='psnr_val', color='orange')
plt.xlabel('Steps (normalized)')
plt.ylabel('PSNR')
plt.legend()
plt.title('PSNR (Average and Validation)')

plt.tight_layout()
plt.show()
```

```python
df['step_normalized'] = df['step'] / 388

# Calculate moving averages
window_size = 5
df['d_loss_avg_ma'] = df['d_loss_avg'].rolling(window=window_size).mean()
df['g_loss_avg_ma'] = df['g_loss_avg'].rolling(window=window_size).mean()
df['psnr_avg_ma'] = df['psnr_avg'].rolling(window=window_size).mean()

df['d_loss_val_ma'] = df['d_loss_val'].rolling(window=window_size).mean()
df['g_loss_val_ma'] = df['g_loss_val'].rolling(window=window_size).mean()
df['psnr_val_ma'] = df['psnr_val'].rolling(window=window_size).mean()

# Plotting
plt.figure(figsize=(15, 15))

# Plot Generator Losses
plt.subplot(3, 1, 1)
plt.plot(df['step_normalized'], df['g_loss_avg'], label='g_loss_avg', color='blue', alpha=0.5)
plt.plot(df['step_normalized'], df['g_loss_val'], label='g_loss_val', color='orange', alpha=0.5)
plt.plot(df['step_normalized'], df['g_loss_avg_ma'], label='g_loss_avg_ma (Moving Avg)', color='blue', linestyle='--')
plt.plot(df['step_normalized'], df['g_loss_val_ma'], label='g_loss_val_ma (Moving Avg)', color='orange', linestyle='--')
plt.xlabel('Steps (normalized)')
plt.ylabel('Generator Loss')
plt.legend()
plt.title('Generator Losses (Average and Validation)')

# Plot Discriminator Losses
plt.subplot(3, 1, 2)
plt.plot(df['step_normalized'], df['d_loss_avg'], label='d_loss_avg', color='blue', alpha=0.5)
plt.plot(df['step_normalized'], df['d_loss_val'], label='d_loss_val', color='orange', alpha=0.5)
plt.plot(df['step_normalized'], df['d_loss_avg_ma'], label='d_loss_avg_ma (Moving Avg)', color='blue', linestyle='--')
plt.plot(df['step_normalized'], df['d_loss_val_ma'], label='d_loss_val_ma (Moving Avg)', color='orange', linestyle='--')
plt.xlabel('Steps (normalized)')
plt.ylabel('Discriminator Loss')
plt.legend()
plt.title('Discriminator Losses (Average and Validation)')

# Plot PSNR
plt.subplot(3, 1, 3)
plt.plot(df['step_normalized'], df['psnr_avg'], label='psnr_avg', color='blue', alpha=0.5)
plt.plot(df['step_normalized'], df['psnr_val'], label='psnr_val', color='orange', alpha=0.5)
plt.plot(df['step_normalized'], df['psnr_avg_ma'], label='psnr_avg_ma (Moving Avg)', color='blue', linestyle='--')
plt.plot(df['step_normalized'], df['psnr_val_ma'], label='psnr_val_ma (Moving Avg)', color='orange', linestyle='--')
plt.xlabel('Steps (normalized)')
plt.ylabel('PSNR')
plt.legend()
plt.title('PSNR (Average and Validation)')

plt.tight_layout()
plt.show()
```
